```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

# Unicode warning 제거 (폰트 관련 경고메시지)
plt.rcParams['axes.unicode_minus']=False

# 한글 폰트 설정
plt.rcParams['font.family'] = "NanumGothic"

# 그래프 출력 사이즈 설정
plt.rcParams["figure.figsize"] = (10, 8)
```

```python
df = pd.read_csv('https://bit.ly/2Vk0zLr')
df.head()
```

```python
df_seoul_year = df_seoul.groupby(['연도','규모']).mean()
df_seoul_year
```

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title ('타이틀', fontsize= 20)

plt.xlabel('x축', fontsize=20)
plt.ylabel('y축', fontsize=20)

plt.xticks(rotation=360)
plt.yticks(rotation=30)

plt.legend([1,2,3],fontsize= 20)


plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2, marker='o',linestyle=':',color='b')
plt.plot(np.arange(10), np.arange(10)**2 ,marker= 'd', linestyle= '-.',color='m')
plt.plot(np.arange(10), np.log(np.arange(10)),marker= 'x', linestyle='--',color='c')

plt.title ('타이틀', fontsize= 20)

plt.xlabel('x축', fontsize=20)
plt.ylabel('y축', fontsize=20)

plt.xticks(rotation=360)
plt.yticks(rotation=30)

plt.legend([1,2,3],fontsize= 20)


plt.xlim(0,5)
plt.ylim(0.5,10)


plt.annotate('dd', xy=(2,4), xytext=(2,5),arrowprops=dict(facecolor='black', shrink=0.05))




plt.grid()


plt.show()
```
