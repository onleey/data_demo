### 데이터 시각화
- 데이터 시각화의 중요성을 강조하고자 할때 영국의 통계학자 프랭크 엔스컴이 만든 앤스컴 콰르텟(앤스컴 콰르텟:Ansombe's quartet)으로 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 함정을 보여주기 위한 그래프이다.
- 앤 스컴 콰르텟 은 4 개의 데이터 셋 으로 구성 되며 각 데이터 셋 에는 2 개의 연속 변수 가 있 습니다 . 4 개 데이터 셋 은 평균 , 분산 , 상관 관계 , 회귀선 이 모두 같다 . 따라서 기술 통계 만 보면마치 같은 데이터 셋 처럼 보일 수 있다 . 하지만 이를 시각화 하면 4 가지 모두 경향 이 다르다는 사실 을 직관적 으로 알 수 있다 . 이런 점 에서 시각화 는 데이터 분석 에서 아주 중요한요소 라고 할 수 있다 .

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
print(type(anscombe))
print(anscombe)
```

```python
import matplotlib.pyplot as plt
dataset_1 = anscombe[anscombe['dataset']=='I']
plt.plot(dataset_1['x'],dataset_1['y'])
plt.show()
```

```python
import matplotlib.pyplot as plt
dataset_1 = anscombe[anscombe['dataset']=='I']

# 선이아닌 점으로 그래프를 그리려면 알파벳 소문자 'o'를 설정 한다.
plt.plot(dataset_1['x'],dataset_1['y'],'o')
plt.show()
```

```python
dataset_1 =anscombe[anscombe['dataset'] =='I']
#print(dataset_1)
dataset_2 =anscombe[anscombe['dataset'] =='II']
#print(dataset_2)
dataset_3 =anscombe[anscombe['dataset'] =='III']
#print(dataset_3)
dataset_4 =anscombe[anscombe['dataset'] =='IV']
print(dataset_4)
```

```python
print(dataset_1.describe())
print(dataset_2.describe())
print(dataset_3.describe())
print(dataset_4.describe())
```

```python
print(dir(dataset_1))
```

```python
# 연산을 하기 위해서 숫자 데이터만 가져온다.
# 연산을 할때 숫자가 아닌 값이 포함되어 있으면 아래와 같은 경고가 발생한다.
#FutureWarning: The default value of numeric_only in DataFrame.var is deprecated. In a future version, it will default to False. In addition, specifying 'numeric_only=None' is deprecated.

dataset_1 = dataset_1[['x','y']]
dataset_2 = dataset_2[['x','y']]
dataset_3 = dataset_3[['x','y']]
dataset_4 = dataset_4[['x','y']]

print(dataset_1)
print(dataset_2)
print(dataset_3)
print(dataset_4)

```

```python
#분산(variance)
print(dataset_1.var())
print(dataset_2.var())
print(dataset_3.var())
print(dataset_4.var())
```

```python
#공분산 (Covariance)
print(dataset_1.cov())
print(dataset_2.cov())
print(dataset_3.cov())
print(dataset_4.cov())
```

```python
#상관계수 (Correlation Coefficient)
print(dataset_1.corr())
print(dataset_2.corr())
print(dataset_3.corr())
print(dataset_4.corr())
```

```python
fig = plt.figure()

#fig.add_subplot(전체 행개수, 전체 열개수, 그래프 순서)

axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)


dataset_1 =anscombe[anscombe['dataset'] =='I']
dataset_2 =anscombe[anscombe['dataset'] =='II']
dataset_3 =anscombe[anscombe['dataset'] =='III']
dataset_4 =anscombe[anscombe['dataset'] =='IV']


axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

#r각각의 그래프마다 제목표시
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')

# 기본틀에 제목표시
fig.suptitle('Anscombe data')

# 그래프의 이름과 숫자가 겹ㅊ쳤을때 각 그래프의 레이아웃을 조절해 준다
fig.tight_layout()




plt.show()
```
